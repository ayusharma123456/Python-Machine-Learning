```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset = pd.read_excel("Online Retail.xlsx")
dataset.head()
```

# Data Clean up of spaces
it involves the data clean up over some of the description so we will drop the rows with no invoice number and credit transition

```python
dataset['Description'] = dataset['Description'].str.strip()
dataset.dropna(axis=0, subset=['InvoiceNo'],inplace=True)
dataset['InvoiceNo']=dataset['InvoiceNo'].astype('str')
dataset=dataset[(~dataset['InvoiceNo'].str.contains('C'))]
dataset
```

# Consolidate
the item into one transition per rows with each product for the sack of keep the dataset small we only  look for sale for friends


```python
basket  = (dataset[dataset['Country'] == "France"]
          .groupby(['InvoiceNo','Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
basket
```

# fun
to keep the positve value converted into one and less than should be 0

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
baskets_sets = basket.applymap(encode_units)
baskets_sets.drop('POSTAGE',inplace=True,axis=1)
baskets_sets
```

# generate a frequent itemset
 that support 7% it use for generate rule for corresponding confidence and lift by apply apriori models with set of association rules

```python
frequent_group = apriori(baskets_sets,min_support=0.07,use_colnames=True)
rules = association_rules(frequent_group,metric="lift",min_threshold=1)
rules.head
```

```python
rules[(rules['lift']>=6)&(rules['confidence']>=0.8)]
```
